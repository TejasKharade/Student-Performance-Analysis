# Data Cleaning

```python
df.isna().sum()
```

```python
df.dropna(subset=['gender'], inplace=True)
```

```python
df
```

```python
df['math score']=df['math score'].fillna(df['math score'].mean())
df['physics score']=df['physics score'].fillna(df['physics score'].mean())
```

```python
df.isna().sum()
```

```python
df['lunch']=df['lunch'].fillna('Unknown')
```

```python
df.isna().sum()
```

Now no Null values are present in the dataframe

```python
df=df.drop(['random_noise','student_id'], axis=1)
```

```python
df
```

# Extracting Some Meaningful Insights

1. Compare Score Distributions by Gender

```python
sns.boxplot(x='gender',y='math score', data=df)
plt.grid(True)

#Comparison Between the math marks scored by male and female
```

```python
sns.boxplot(x='gender',y='physics score', data=df)
plt.grid(True)
```

Can we simultaneously compare the marks of physics and maths????

2. Impact of Lunch Type on Performance

```python
sns.boxplot(x='lunch',y='math score',data=df)
plt.grid(True)
```

```python
sns.boxplot(x='lunch',y='physics score',data=df)
plt.grid(True)
```

3. Parental Education vs Student Performance

Does having more educated parents help?

```python
sns.boxplot(x='parental level of education', y='math score', data=df)
plt.grid(True)
plt.xticks(rotation=45)
```

4. Comparison between the marks of differnt subjects

```python
plt.figure(figsize=(10,6))
sns.pairplot(df[['math score','physics score','chemistry score','gender']], hue='gender')
```

Simultaneously comparing the scores of each subject

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.barplot(x='gender', y='math score', data=df, ax=axes[0])
axes[0].set_title("Math Score")

sns.barplot(x='gender', y='physics score', data=df, ax=axes[1])
axes[1].set_title("physics Score")

sns.barplot(x='gender', y='chemistry score', data=df, ax=axes[2])
axes[2].set_title("chemistry Score")

```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.barplot(x='test preparation course', y='math score', data=df, ax=axes[0])
axes[0].set_title("Math Score")

sns.barplot(x='test preparation course', y='physics score', data=df, ax=axes[1])
axes[1].set_title("physics Score")

sns.barplot(x='test preparation course', y='chemistry score', data=df, ax=axes[2])
axes[2].set_title("chemistry Score")

fig.suptitle('Score vs Test Preparation')
```
